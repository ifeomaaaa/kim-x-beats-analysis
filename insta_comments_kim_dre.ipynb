```python
import pandas as pd 
df=pd.read_csv('IGCommentsExport_ChCpQlzr8_a_503_comments.csv', encoding='utf8')
```

```python
df.head(100)
```

```python
from textblob import TextBlob
```

```python
def getTextSubjectivity(txt):
  return TextBlob(txt).sentiment.subjectivity

def getTextPolarity(txt):
  return TextBlob(txt).sentiment.polarity
```

```python
df['Subjectivity']=df['Comment Text'].apply(getTextSubjectivity)
df['Polarity']=df['Comment Text'].apply(getTextPolarity) #polarity<0 = -ve sentiment, =0 =neutral and >0 = +ve
df.head(2)
```

```python
def getTextAnalysis(a):
  if a<0:
    return "Negative"
  elif a==0:
    return "Neutral"
  else:
    return "Positive"
```

```python
df["Score"]=df['Polarity'].apply(getTextAnalysis)
df.head(5)
```

```python
positive=df[df['Score']=="Positive"]
print(str(positive.shape[0]/(df.shape[0])*100)+"% of Positive tweets")
pos=positive.shape[0]/df.shape[0]*100
```

```python
negative=df[df['Score']=="Negative"]
print(str(negative.shape[0]/(df.shape[0])*100)+"% of Negative tweets")
neg=negative.shape[0]/df.shape[0]*100
```

```python
neutral=df[df['Score']=="Neutral"]
print(str(neutral.shape[0]/(df.shape[0])*100)+"% of Neutral tweets")
neu=neutral.shape[0]/df.shape[0]*100
```
